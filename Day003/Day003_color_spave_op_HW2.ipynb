```python
import cv2
import numpy as np
```

```python
lena_ori = cv2.imread("image/lena.png")
cv2.imshow("Original", lena_ori)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### change saturation
HLS color space －> change s channel

```python
# [Hue, Lightness, Saturation]
lena_hls = cv2.cvtColor(lena_ori, cv2.COLOR_BGR2HLS)
lena_hls
```

```python
lena_hls[:,:,-1]
```

```python
change_percentage = 0.2

lena_hls_down = lena_hls.astype('float32')
lena_hls_down[:,:, -1] = lena_hls_down[:,:, -1]/255 - change_percentage
lena_hls_down[lena_hls_down[:,:, -1] < 0] = 0
lena_hls_down[:,:, -1] = np.round(lena_hls_down[:,:, -1] * 255)
lena_hls_down
```

```python
lena_hls_up = lena_hls.astype('float32')
lena_hls_up[..., -1] = lena_hls_up[..., -1]/255 + change_percentage
lena_hls_up[lena_hls_up[..., -1] > 1] = 1
lena_hls_up[..., -1] = np.round(lena_hls_up[:,:, -1] * 255)
lena_hls_up
```

```python
lena_bgr_down = cv2.cvtColor(lena_hls_down, cv2.COLOR_HLS2BGR)
lena_bgr_up = cv2.cvtColor(lena_hls_up, cv2.COLOR_HLS2BGR)
```

```python
lena_change = np.hstack((lena_hls, lena_bgr_down, lena_bgr_up))
lena_change
```

```python
cv2.imshow("change saturation", lena_change)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
